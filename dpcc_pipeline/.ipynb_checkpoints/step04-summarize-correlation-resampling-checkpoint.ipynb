```python
def ut_as_list( dframe, diag=1, cols=['Row','Column','Value'] ):
  """
  for a symmetric dataframe, where cols=rows, get the upper triangle as a list of row/column pairs
  diag = 1 (default): ignore diagonal
  diag = 0: include diagonal
  """
  #if (dframe.index.name == dframe.columns.name):
  dframe.index.name = cols[0]
  dframe.columns.name = cols[1]
  #             dframe.index.name = dframe.index.name + '.1'
  #             dframe.index.name = dframe.index.name + '.2'
  d = dframe.where( np.triu( np.ones( dframe.shape ), k=diag).astype(np.bool))
  d = d.stack().reset_index()
  d.columns=cols
  return d
```

```python
cc_aml = pd.read_table('cc_matrix_aml15cells_17407genes.txt', sep=',', index_col=0)
cc_exBld=pd.read_table('cc_matrix_exBlood530cells_17407genes.txt', sep=',', index_col=0)
```

```python
ccAmlList = ut_as_list(cc_aml, cols=['G1','G2','PCC_AML'])
ccOthList = ut_as_list(cc_exBld, cols=['G1','G2','PCC_Other'])
ccPvalList= ut_as_list(cc_count_gt, cols=['G1','G2','Pval'])
```

```python
ccAmlList = ccAmlList.join( ccOthList['PCC_Other'], how='inner')
ccAmlList['dPCC'] = ccAmlList.PCC_AML - ccAmlList.PCC_Other
ccAmlList = ccAmlList.join( ccPvalList['Pval'], how='inner')
```

```python
ccAmlList.sort_values(['Pval','dPCC'], ascending=[True,False], inplace=True)
```
