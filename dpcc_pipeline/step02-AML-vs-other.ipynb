```python
z = pd.read_csv('./zscore_550cells_min9perLineage.txt', index_col=0, sep='\t')
z.info()
```

```python
cells = pd.read_csv('./sample_info_550cells.txt', index_col=0, sep='\t')
cells.info()
```

```python
cells.groupby(['lineage','lineage_subtype']).count()
```

```python
aml = [x for x in cells.index if cells.loc[x]['lineage_subtype']=='AML']
cells.loc[aml]
```

```python
cc_aml = z.loc[aml].corr()
```

```python
cc_aml.to_csv('cc_matrix_aml15cells_17407genes.txt', sep=',', float_format='%4.3f')
```

```python
ex_blood = [x for x in cells.index if cells.loc[x]['lineage']!='blood']
len(ex_blood)
```

```python
cc_ex_blood = z.loc[ex_blood].corr()
```

```python
cc_ex_blood.to_csv('cc_matrix_exBlood530cells_17407genes.txt', sep=',', float_format='%4.3f')
```

```python
cc_delta = cc_aml - cc_ex_blood
```

```python
cc_delta.to_csv('delta-cc-aml-minus-exBlood.txt', sep=',', float_format='%4.3f', index=True)
```
